```python
def remove_rare_categories(df, threshold=0.01):
    # Loop through each categorical column (dtype 'object') in the DataFrame
    for col in df.select_dtypes(include='object').columns:
        # Calculate normalized frequency (proportion) of each category
        freq = df[col].value_counts(normalize=True)
        # Identify categories that occur less than the specified threshold
        rare_labels = freq[freq < threshold].index
        # Remove rows where the column contains any of the rare categories
        df = df[~df[col].isin(rare_labels)]
    
    # Return the cleaned DataFrame with reset index
    return df.reset_index(drop=True)

# Apply the function to the DataFrame 'data' to remove rare categories
data = remove_rare_categories(data)

# Check the shape of the resulting DataFrame
data.shape
```

```python
# Log transform 'capital-gain' and 'capital-loss'
import numpy as np
data['capital-gain'] = data['capital-gain'].apply(lambda x: np.log1p(x))
data['capital-loss'] = data['capital-loss'].apply(lambda x: np.log1p(x))

# capital-gain category visualization
plt.boxplot(data['capital-gain'])
plt.show() 

# capital-loss category visualization
plt.boxplot(data['capital-loss'])
plt.show()

# What it does:
# - np.log1p(x) computes the natural logarithm of x + 1. It’s safer than np.log(x) when x = 0 because log(0) is undefined.
# - This transformation compresses the range of values, especially when the original data has large disparities.

# Why it's useful:
# - Capital gains and losses can have very skewed distributions—many zero values and a few very large ones.
# - Log transformation:
# - Helps normalize the data.
# - Makes patterns easier to model.
# - Improves the performance of many machine learning algorithms by stabilizing variance and reducing the effect of outliers.
```

```python
# Cap 'hours-per-week' at 80
data['hours-per-week'] = data['hours-per-week'].apply(lambda x: min(x, 80))

# hours-per-week category visualization
plt.boxplot(data['hours-per-week'])
plt.show() 

#  What it does:
# - This line goes through each value in the hours-per-week column.
# - If a person is listed as working more than 80 hours, their value gets capped at 80.
# - In other words, if x > 80, it becomes 80. If x <= 80, it stays the same.
#  Why it's useful:
# - It prevents extremely high values from skewing your analysis.
# - Keeps outliers in check, especially if you're working with models like regression or visualization—outliers can mess with interpretations and influence predictions disproportionately.
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from xgboost import XGBClassifier
```

```python
# Create a dictionary for defining models and tuning hyperparameters
# Models and parameters
model_params = {
    'LogisticRegression': {
        'model': LogisticRegression(max_iter=1000),
        'params': {
            'C': [0.1, 1, 10],
            'solver': ['liblinear', 'lbfgs']
        }
    },
    'DecisionTree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'max_depth': [3, 5, 10],
            'min_samples_split': [2, 5, 10]
        }
    },
    'RandomForest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [50, 100],
            'max_depth': [5, 10]
        }
    },
    'XGBoost': {
        'model': XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
        'params': {
            'n_estimators': [50, 100],
            'max_depth': [3, 5],
            'learning_rate': [0.05, 0.1]
        }
    },
    'GradientBoosting': {
        'model': GradientBoostingClassifier(),
        'params': {
            'n_estimators': [50, 100],
            'learning_rate': [0.05, 0.1],
            'max_depth': [3, 5]
        }
    },
    'SVM': {
        'model': SVC(),
        'params': {
            'C': [0.1, 1],
            'kernel': ['linear', 'rbf']
        }
    },
    'KNN': {
        'model': KNeighborsClassifier(),
        'params': {
            'n_neighbors': [3, 5, 7]
        }
    }
}
```

```python
results = []
best_accuracy = 0
best_model = None
best_model_name = ""
best_params = {}

for name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=3, scoring='accuracy', n_jobs=-1)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)

    print(f"\n{name} Accuracy: {acc:.4f}")
    print("Best Params:", clf.best_params_)
    print("Classification Report:\n", classification_report(y_test, y_pred))

    results.append({
        'Model': name,
        'Best Params': clf.best_params_,
        'Test Accuracy': acc
    })

    # Save best model
    if acc > best_accuracy:
        best_accuracy = acc
        best_model = clf.best_estimator_
        best_model_name = name
        best_params = clf.best_params_

# Show sorted results
results_df = pd.DataFrame(results)
print("\nModel Comparison:")
print(results_df.sort_values(by='Test Accuracy', ascending=False))
```

```python
# Save the best model
import joblib
joblib.dump(best_model, "best_model.pkl")
print(f"\n✅ Best model: {best_model_name} with accuracy {best_accuracy:.4f}")
print("✅ Saved best model as best_model.pkl")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
import sklearn # Ensure scikit-learn is imported

# --- Page Configuration ---
st.set_page_config(
    page_title="Salary Predictor Pro",
    page_icon="💰",
    layout="wide",
    initial_sidebar_state="expanded"
)

# --- Caching Models ---
# Using st.cache_resource to load the model and preprocessor only once
@st.cache_resource
def load_model_and_preprocessor():
    """Loads the trained model and the preprocessor from disk."""
    try:
        model = joblib.load("best_model.pkl")
        preprocessor = joblib.load("preprocessor.pkl")
        return model, preprocessor
    except FileNotFoundError:
        st.error("Model or preprocessor not found. Please ensure 'best_model.pkl' and 'preprocessor.pkl' are in the correct directory.")
        return None, None

model, preprocessor = load_model_and_preprocessor()

# --- Helper Function for Prediction ---
def make_prediction(input_df):
    """Preprocesses input and returns prediction and probability."""
    if model is None or preprocessor is None:
        return None, None
    
    # The preprocessor expects a DataFrame.
    processed_input = preprocessor.transform(input_df)
    prediction = model.predict(processed_input)
    probability = model.predict_proba(processed_input)
    return prediction[0], probability[0]

# --- Sidebar for User Inputs ---
with st.sidebar:
    st.header("📊 Input Employee Data")
    st.markdown("Enter the details of the employee to predict their salary class.")

    # Using columns for a cleaner layout
    col1, col2 = st.columns(2)
    with col1:
        age = st.number_input("Age", 17, 70, 35)
    with col2:
        hours_per_week = st.number_input("Hours per Week", 1, 80, 40)

    education = st.selectbox("Education Level", [
        "Bachelors", "HS-grad", "Some-college", "Masters", "Assoc-voc",
        "Assoc-acdm", "Doctorate", "Prof-school"
    ])
    
    occupation = st.selectbox("Occupation", [
        "Prof-specialty", "Exec-managerial", "Craft-repair", "Adm-clerical",
        "Sales", "Other-service", "Machine-op-inspct", "Transport-moving",
        "Handlers-cleaners", "Farming-fishing", "Tech-support", "Protective-serv",
        "Priv-house-serv", "Armed-Forces"
    ])
    
    # Add other features your model was trained on
    # Example:
    # capital_gain = st.number_input("Capital Gain", 0, 99999, 0)
    # marital_status = st.selectbox("Marital Status", ["Married-civ-spouse", "Never-married", ...])

# --- Main Page Content ---
st.title("👩‍💻 Employee Salary Predictor👨‍💻")
st.markdown("Predict if an employee earns more than **50K** or less than equal **50k** a year based on their profile. This app uses a machine learning model to make predictions.")

# Create the input DataFrame from user selections
# IMPORTANT: The column names must match exactly what the preprocessor expects.
input_data = {
    'age': [age],
    'education': [education],
    'occupation': [occupation],
    'hours-per-week': [hours_per_week]
    # Add other features here, matching the training columns
    # 'capital-gain': [capital_gain],
    # 'marital-status': [marital_status]
}
input_df = pd.DataFrame(input_data)

st.write("---")

# --- Prediction Display ---
if model is not None:
    if st.button("🚀 Predict Salary Class", type="primary"):
        prediction, probability = make_prediction(input_df)

        st.subheader("🔮 Prediction Result")
        
        # Displaying the result in a more visual way
        if prediction == ">50K":
            st.success(f"**Prediction: High Salary (>50K)**", icon="✅")
        else:
            st.info(f"**Prediction: Standard Salary (≤50K)**", icon="☑️")
        
        # Displaying probability with a progress bar
        salary_class_index = list(model.classes_).index(prediction)
        prob_value = probability[salary_class_index]
        st.write(f"**Confidence:**")
        st.progress(prob_value, text=f"{prob_value:.2%}")

        # Expander for showing more details
        with st.expander("Show Input Data and Probabilities"):
            st.write("Your Input Data:")
            st.dataframe(input_df)
            st.write("Prediction Probabilities:")
            prob_df = pd.DataFrame([probability], columns=model.classes_)
            st.dataframe(prob_df)

# --- Batch Prediction Section ---
st.write("---")
st.header("📂 Batch Prediction from CSV")

uploaded_file = st.file_uploader("Upload a CSV file for batch predictions.", type="csv")

if uploaded_file is not None and preprocessor is not None:
    batch_df = pd.read_csv(uploaded_file)
    
    # Check if required columns are in the uploaded file
    required_cols = preprocessor.feature_names_in_
    if not all(col in batch_df.columns for col in required_cols):
        st.error(f"Error: The uploaded CSV must contain the following columns: {', '.join(required_cols)}")
    else:
        st.write("Uploaded Data Preview:")
        st.dataframe(batch_df.head())

        # Preprocess and predict
        processed_batch = preprocessor.transform(batch_df[required_cols])
        batch_predictions = model.predict(processed_batch)
        
        # Add predictions to the DataFrame
        results_df = batch_df.copy()
        results_df['Predicted_Salary'] = batch_predictions
        
        st.success("Predictions generated successfully!", icon="🎉")
        st.dataframe(results_df.head())
        
        # Provide a download button for the results
        csv = results_df.to_csv(index=False).encode('utf-8')
        st.download_button(
            label="⬇️ Download Predictions",
            data=csv,
            file_name='predicted_salaries.csv',
            mime='text/csv',
        )
```
